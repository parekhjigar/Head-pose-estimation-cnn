## Building model

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Activation
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Lambda
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Input
import tensorflow as tf
```

## Function to build tilt branch of model

```python
def build_tilt_branch(inputs, numTilts, finalAct="softmax", chanDim=-1):
    # utilize a lambda layer to convert the 3 channel input to a
    # grayscale representation
    x = Lambda(lambda c: tf.image.rgb_to_grayscale(c))(inputs)
    # CONV => RELU => POOL
    x = Conv2D(32, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(3, 3))(x)
    x = Dropout(0.25)(x)


    # (CONV => RELU) * 2 => POOL
    x = Conv2D(64, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Conv2D(64, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Dropout(0.25)(x)

    # (CONV => RELU) * 2 => POOL
    x = Conv2D(128, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Conv2D(128, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Dropout(0.25)(x)

    # define a branch of output layers for the number of different tilts
    x = Flatten()(x)
    x = Dense(256)(x)
    x = Activation("relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(numTilts)(x)
    x = Activation(finalAct, name="tilt_output")(x)

    # return the tilt prediction sub-network
    return x
```

## Function to build pan branch of model

```python
def build_pan_branch(inputs, numPans, finalAct="softmax", chanDim=-1):
    # utilize a lambda layer to convert the 3 channel input to a
    # grayscale representation
    x = Lambda(lambda c: tf.image.rgb_to_grayscale(c))(inputs)
    # CONV => RELU => POOL
    x = Conv2D(32, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(3, 3))(x)
    x = Dropout(0.25)(x)


    # (CONV => RELU) * 2 => POOL
    x = Conv2D(64, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Conv2D(64, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Dropout(0.25)(x)

    # (CONV => RELU) * 2 => POOL
    x = Conv2D(128, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = Conv2D(128, (3, 3), padding="same")(x)
    x = Activation("relu")(x)
    x = BatchNormalization(axis=chanDim)(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Dropout(0.25)(x)

    # define a branch of output layers for the number of different tilts
    x = Flatten()(x)
    x = Dense(256)(x)
    x = Activation("relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(numPans)(x)
    x = Activation(finalAct, name="pan_output")(x)

    # return the pan prediction sub-network
    return x
```

## Function to build final model out of the sub-networks

```python
def build(width, height, numTilts, numPans,
    finalAct="softmax"):
    # initialize the input shape and channel dimension (this code
    # assumes you are using TensorFlow which utilizes channels
    # last ordering)
    inputShape = (height, width, 3)
    chanDim = -1

    # construct both the "tilt" and "pan" sub-networks
    inputs = Input(shape=inputShape)
    tiltBranch = build_tilt_branch(inputs, numTilts, finalAct=finalAct, chanDim=chanDim)
    panBranch = build_pan_branch(inputs, numPans, finalAct=finalAct, chanDim=chanDim)

    # create the model using our input (the batch of images) and
    # two separate outputs -- one for the clothing tilt
    # branch and another for the pan branch, respectively
    model = Model(
    inputs=inputs,
    outputs=[tiltBranch, panBranch],
    name="headpose")

    # return the constructed network architecture
    return model
```

## Model

```python
# initialize the number of epochs to train for, initial learning rate,
# batch size, and image dimensions
EPOCHS = 50
INIT_LR = 1e-3
BS = 32
IMGWIDTH = 144
IMGHEIGHT = 192

model = build(IMGWIDTH, IMGHEIGHT, numTilts=len(tiltLB.classes_), numPans=len(panLB.classes_), finalAct="softmax")
```

## Losses

```python
# define two dictionaries: one that specifies the loss method for
# each output of the network along with a second dictionary that
# specifies the weight per loss
losses = {
    "tilt_output": "categorical_crossentropy",
    "pan_output": "categorical_crossentropy",
}
lossWeights = {"tilt_output": 1.0, "pan_output": 1.0}
```

## Initialize the optimizer and compile the model

```python
opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
model.compile(optimizer=opt, loss=losses, loss_weights=lossWeights, metrics=["accuracy"])
```

## Train the network to perform multi-output classification

```python
H = model.fit(x=trainX,
              y={"tilt_output": trainTiltY, "pan_output": trainPanY},
              validation_data=(testX, {"tilt_output": testTiltY, "pan_output": testPanY}),
              epochs=EPOCHS,
              verbose=1)
```
